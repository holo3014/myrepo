## 第3問 クラス（5点）
名前と血液型と体重と身長をインスタンス変数とし、BMIを計算をするcalc_bmiメソッドと、指標に基づいて体型を評価するeval_bmiメソッドと、情報を表示するshowメソッドをもつクラス「Person」を作成したい。BMIは体重 $w[kg]$, 身長$h[m]$とすると、$$BMI = w/h^2$$と表すことができる。また、$$18.5 \leq BMI < 25.0$$を標準BMIとし、標準BMI未満は痩せ型、標準BMI以上は肥満型とする。



ここでは、例えば、名前が○○, 血液型が?型, 体重45.0kg, 身長1.50mの人は
```
○○さんの血液型は?型で、BMIは20.00で標準体型です。
```
と表示されるようにしたい。  
以下のコードを埋めてください。

````python
class Person:
    def __init__(self, name, blood_type, weight, height):
        # ここに入力してください。（1点）
        self.name = name
        self.blood_type = blood_type
        self.weight = weight
        self.height = height
        
    
    def calc_bmi(self):
        # ここに入力してください。（1点）
        return self.weight / (self.height**2)
        
    
    def eval_bmi(self):
        # ここに入力してください。（1点）
        b = self.bmi()
        if b>=25:
            s = '肥満型'
        elif b>=18.5:
            s = '標準体型'
        else:
            s = '痩せ型'
        return s
        
    def show(self):
        # 上の例のように出力(BMIは小数点以下2桁まで表示)。
        # ここに入力してください。（1点）
        print('{}さんの血液型は{}型で、BMIは{}で{}です。BMIは{:.1f}で'.format(self.name(),self.blood_type(),self.calc_bmi(),self.eval_bmi())

# 人名が「河内」、血液型は「O型」、体重は「60.0kg」、 身長は「1.75m」インスタンスを生成して、show()で表示させてください。（1点）
````
